# Intro to Fourier transform
## Part 1
### A

```python
import numpy
from matplotlib import pyplot as plot

x = numpy.linspace(0,2,500)
y = 2 * numpy.sin(2 * numpy.pi * 3 * x + (numpy.pi / 3))
plot.plot(x,y,color='c')
plot.title('Sine Wave with Amplitude = 2, Frequency = 3 Hz, Phase = +3/PI and Sampling rate of 500')
plot.ylabel('Amplitude')
plot.xlabel('Time')
plot.grid()
plot.show()
```

### B

```python
y = 2 * numpy.exp((0.0 + 1.j) * 2 * numpy.pi * x * 3 + (numpy.pi / 3))
plot.plot(x,y.real,color='r',label='Real')
plot.plot(x,y.imag,color='b',linestyle='dashed',label='Imaginary')
plot.title('Sine Wave')
plot.ylabel('Amplitude')
plot.xlabel('Time')
plot.legend()
plot.grid()
plot.show()
```

### C

```python
fig = plot.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot(x, y.real, y.imag, color='c')
ax.set_title("3D Complex Sine Wave")
ax.set_xlabel("Time")
ax.set_ylabel("Real")
ax.set_zlabel("Imaginary")
ax.grid()
plot.show()
```
