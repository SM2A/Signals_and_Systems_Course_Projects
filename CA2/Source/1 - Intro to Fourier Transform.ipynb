# Intro to Fourier transform
## Part 1
### A

```python
import numpy
from matplotlib import pyplot as plot

x = numpy.arange(0, 2, 1./500)
y = 2 * numpy.sin(2 * numpy.pi * 3 * x + (numpy.pi / 3))
plot.plot(x,y,color='c')
plot.title('Sine Wave with Amplitude = 2, Frequency = 3 Hz, Phase = $\\frac{\pi}{3}$ and Sampling rate of 500')
plot.ylabel('Amplitude')
plot.xlabel('Time')
plot.grid()
plot.show()
```

### B

```python
y = 2 * numpy.exp((0.0 + 1.j) * 2 * numpy.pi * x * 3 + (numpy.pi / 3))
plot.plot(x,y.real,color='r',label='Real')
plot.plot(x,y.imag,color='b',linestyle='dashed',label='Imaginary')
plot.title('Sine Wave')
plot.ylabel('Amplitude')
plot.xlabel('Time')
plot.legend()
plot.grid()
plot.show()
```

### C

```python
fig = plot.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot(x, y.real, y.imag, color='c')
ax.set_title("3D Complex Sine Wave")
ax.set_xlabel("Time")
ax.set_ylabel("Real")
ax.set_zlabel("Imaginary")
ax.grid()
plot.show()
```

## Part 2
### A

The dot product is a single that reflects the relation between two vectors. The dot product is the computational backbone
of many algorithms and mathematical operations, including the Fourier transform, convolution, and correlation. To compute
the dot product between two vectors, simply point-wise multiply the corresponding elements and then sum over all elements.
The larger the dot product, the more similar the two vectors are two each other.

### B

#### 1

```python
wave1 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2))
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')
```

#### 2

When two sine waves are the same we have a very large dot product. But as soon as we change the frequency, we are making
them less similar so we get smaller results in our dot product.

```python
wave1 = 2 * numpy.sin( 2 * numpy.pi * 5.5 * x + (numpy.pi / 2))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 4.5 * x + (numpy.pi / 2))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 5.35 * x + (numpy.pi / 2))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 4.65 * x + (numpy.pi / 2))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')
```

#### 3

Same as previous part we are making two signals less identical.

```python
wave1 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + numpy.pi)
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 5 * x)
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (5 * numpy.pi / 6))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')

wave1 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 6))
wave2 = 2 * numpy.sin( 2 * numpy.pi * 5 * x + (numpy.pi / 2) )
dp = numpy.dot(wave1, wave2)
print(f'Dot Product = {dp}')
```

### C

#### 1

```python
x = numpy.arange(-1, 1, 1./1000)
y = numpy.sin(2 * numpy.pi * 5 * x) * numpy.exp((-1) * (x**2))
plot.plot(x, y, color='c')
plot.title('Signal')
plot.ylabel('Amplitude')
plot.xlabel('Time')
plot.grid()
plot.show()
```

#### 2

As we can see in 5 Hz we have maximum dot product because it's identical to our sine wave.

```python
freq = numpy.arange(2, 10, 0.5)
wave = [numpy.dot(y, numpy.sin(2 * numpy.pi * f * x) * numpy.exp((-1) * (x**2))) for f in freq]
plot.stem(freq, wave)
plot.title('Dot Product')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()
plot.show()
```

#### 3

Still we have maximum dot product in 5 Hz but much less than prevous part because we are shifting our sine wave.

```python
y = numpy.sin(2 * numpy.pi * 5 * x + (numpy.pi / 2)) * numpy.exp((-1) * (x**2))
wave = [numpy.dot(y, numpy.sin(2 * numpy.pi * f * x) * numpy.exp((-1) * (x**2))) for f in freq]
plot.stem(freq, wave)
plot.title('Dot Product $\\frac{\pi}{2}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()
plot.show()
```

```python
y = numpy.sin(2 * numpy.pi * 5 * x + (numpy.pi / 4)) * numpy.exp((-1) * (x**2))
wave = [numpy.dot(y, numpy.sin(2 * numpy.pi * f * x) * numpy.exp((-1) * (x**2))) for f in freq]
plot.stem(freq, wave)
plot.title('Dot Product $\\frac{\pi}{4}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()
plot.show()
```

#### 1

```python
y = numpy.sin(2 * numpy.pi * 5 * x + (numpy.pi / 2)) * numpy.exp((-1) * (x**2))
wave = [numpy.vdot(y, numpy.exp(1.j * 2 * numpy.pi * f * x)) for f in freq]

plot.figure(figsize=(10,4))

plot.subplot(1,2,1)
plot.stem(freq, numpy.real(wave))
plot.title('Real $\\frac{\pi}{2}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()

plot.subplot(1,2,2)
plot.stem(freq, numpy.imag(wave))
plot.title('Imaginary $\\frac{\pi}{2}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()

plot.tight_layout()
plot.show()
```

```python
y = numpy.sin(2 * numpy.pi * 5 * x + (numpy.pi / 4)) * numpy.exp((-1) * (x**2))
wave = [numpy.vdot(y, numpy.exp(1.j * 2 * numpy.pi * f * x)) for f in freq]

plot.figure(figsize=(10,4))

plot.subplot(1,2,1)
plot.stem(freq, numpy.real(wave))
plot.title('Real $\\frac{\pi}{4}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()

plot.subplot(1,2,2)
plot.stem(freq, numpy.imag(wave))
plot.title('Imaginary $\\frac{\pi}{4}$')
plot.ylabel('Dot Product')
plot.xlabel('Frequency (Hz)')
plot.grid()

plot.tight_layout()
plot.show()
```
