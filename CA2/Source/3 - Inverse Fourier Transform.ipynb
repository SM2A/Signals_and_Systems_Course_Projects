# Inverse Fast Fourier Transform / Discrete Inverse Fourier Transform

## C

Because plotting all process is taking to much space, just final step is plotted.

```python
import numpy
from matplotlib import pyplot as plot

time  = numpy.arange(0,3,1/1000)
signal  = (1 + numpy.sin(2 * numpy.pi * 12 * time)) * (numpy.cos(numpy.sin(2 * numpy.pi * 25 * time) + time))

ft = numpy.array(range(len(time))) / len(time)
fc = numpy.zeros(len(signal), dtype=complex)

for f in range(len(time)):
    cs = numpy.exp(-1j * 2 * numpy.pi * f * ft)
    fc[f] = numpy.sum(signal * cs) / len(time)

freq = numpy.linspace(0,1000,num=len(time))
rs = numpy.zeros(len(signal),dtype=complex)

# for f in range(len(time)):

#     cs = fc[f] * numpy.exp(1j * 2 * numpy.pi * f * ft)
#     rs = rs + cs


#     plot.figure(figsize=(15,4))

#     plot.subplot(1,2,1)
#     plot.plot(time,signal,label='Original',color='r')
#     plot.plot(time,numpy.real(rs),label='Reconstruction',color='b')
#     plot.legend()

#     plot.subplot(1,2,2)
#     plot.plot(freq[0:f],2 * numpy.abs(fc[0:f]))

#     plot.show()

for f in range(len(time)):

    cs = fc[f] * numpy.exp(1j * 2 * numpy.pi * f * ft)
    rs = rs + cs


plot.figure(figsize=(15,4))

plot.subplot(1,2,1)
plot.plot(time,signal,label='Original',color='r')
plot.plot(time,numpy.real(rs),label='Reconstruction',color='b')
plot.legend()

plot.subplot(1,2,2)
plot.plot(freq[0:f],2 * numpy.abs(fc[0:f]))

plot.show()
```
