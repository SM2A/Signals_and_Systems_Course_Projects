# Fast Fourier Transform / Discrete Fourier Transform

```python
import numpy
from matplotlib import pyplot as plot

y = []
x = numpy.linspace(0, 2, 2000)

with open('../Data/Signal.txt') as file_in:
    for line in file_in:
        y.append(float(line))

plot.plot(x, y, color='c')
plot.grid()
plot.show()
```
