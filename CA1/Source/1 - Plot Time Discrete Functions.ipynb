# Question 1

Discrete time function $x[n] = A cos[2 \pi kn]$ with arbitrary A and $k \in (0,1)$

```python
import numpy
from matplotlib import pyplot

T = 1
k = 0.25
A = 1
border = numpy.pi

def cos_2_pi_k_n(n):
    return A*numpy.cos(2*numpy.pi*k*n)

def draw_plot(x_,y_):
    pyplot.stem(x_,y_)
    pyplot.ylabel('y[n]')
    pyplot.xlabel('n')
    pyplot.grid()
    pyplot.show()

n = numpy.linspace(-border,border,50)
```

### Part A

Assuming $T$ is sampling period we have :

A. $y[n] = \frac{1}{T}(x[n+1]-x[n])$

```python
draw_plot(n,(1/T)*(cos_2_pi_k_n(n+1)-cos_2_pi_k_n(n)))
```

B. $y[n] = \frac{1}{2T}(x[n+1]-x[n-1])$

```python
draw_plot(n,(1/(2*T))*(cos_2_pi_k_n(n+1)-cos_2_pi_k_n(n-1)))
```

C. $y[n] = \frac{1}{T^2}(x[n+1]-2x[n]+x[n-1])$

```python
draw_plot(n,(1/(T**2))*(cos_2_pi_k_n(n+1)-(2*cos_2_pi_k_n(n))+cos_2_pi_k_n(n-1)))
```

D. $y[n] = 0.1x[n-2]+0.2x[n-1]+0.4x[n]+0.2x[n+1]+0.1x[n+2]$

```python
draw_plot(n,
          (0.1*cos_2_pi_k_n(n-2))+(0.2*cos_2_pi_k_n(n-1))+cos_2_pi_k_n(n)
          +(0.2*cos_2_pi_k_n(n+1))+(0.1*cos_2_pi_k_n(n+2)))
```
