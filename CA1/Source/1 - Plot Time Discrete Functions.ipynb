# Question 1

Discrete time function $x[n] = A cos[2 \pi kn]$ with arbitrary A and $k \in (0,1)$

```python
import numpy
from matplotlib import pyplot

T = 100
k = 0.5
A = 1

def cos_2_pi_k_n(n):
    return A*numpy.cos(2*numpy.pi*k*n)
```

### Part A

Assuming $T$ is sampling period we have :

A. $y[n] = \frac{1}{T}(x[n+1]-x[n])$

```python
n = numpy.arange(-5,5,0.1)
y = (1/T)*(cos_2_pi_k_n(n+1)-cos_2_pi_k_n(n))
pyplot.stem(n,y)
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.show()
```

B. $y[n] = \frac{1}{2T}(x[n+1]-x[n-1])$

```python
n = numpy.arange(-5,5,0.1)
y = (1/(2*T))*(cos_2_pi_k_n(n+1)-cos_2_pi_k_n(n-1))
pyplot.stem(n,y)
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.show()
```

C. $y[n] = \frac{1}{T^2}(x[n+1]-2x[n]+x[n-1])$

```python
n = numpy.arange(-5,5,0.1)
y = (1/(T**2))*(cos_2_pi_k_n(n+1)-(2*cos_2_pi_k_n(n))+cos_2_pi_k_n(n-1))
pyplot.stem(n,y)
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.show()
```

D. $y[n] = 0.1x[n-2]+0.2x[n-1]+0.4x[n]+0.2x[n+1]+0.1x[n+2]$

```python
n = numpy.arange(-5,5,0.1)
y = (0.1*cos_2_pi_k_n(n-2))+(0.2*cos_2_pi_k_n(n-1))+cos_2_pi_k_n(n)+(0.2*cos_2_pi_k_n(n+1))+(0.1*cos_2_pi_k_n(n+2))
pyplot.stem(n,y)
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.show()
```
