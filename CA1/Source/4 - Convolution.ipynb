## Discrete time

### Part A

Numpy convolution returns the discrete, linear convolution of two one-dimensional sequences but Scipy convolution
convolves two N-dimensional arrays.

```python
import numpy
from scipy import signal
from matplotlib import pyplot

t = numpy.linspace(-50, 50, 100000)

alpha = 2

x = numpy.exp((-1) * t * alpha) * numpy.heaviside(t, 1)
h = numpy.heaviside(t, 1)

y = signal.convolve(x, h, mode='same')
pyplot.plot(t, y)
pyplot.title('Numpy')
pyplot.xlabel('n')
pyplot.ylabel('convolution')
pyplot.grid()
pyplot.show()

y = numpy.convolve(x, y, mode='same')
pyplot.plot(t, y)
pyplot.title('Scipy')
pyplot.xlabel('n')
pyplot.ylabel('convolution')
pyplot.grid()
pyplot.show()
```
