### Part A

We can divide n into 3 intervals. $n < 0$, $0 \leqslant n \leqslant4$ , and $n > 4$ .

For each interval we have :

**$n < 0$** : $x[n] \circledast h[n] = $

**$0 \leqslant n \leqslant4$** : $x[n] \circledast h[n] = \sum_{k=0}^{n} x[n]h[n-k]$

**$n > 4$** : $x[n] \circledast h[n] = \sum_{k=0}^{4} x[n]h[n-k]$

### Part B

```python
import numpy
from scipy import signal
from matplotlib import pyplot
```

Because for $n < 0$ convolution is *zero* we will not calculate and plot $n < 0$ interval.

$0 \leqslant n \leqslant4$

```python
n = [j for j in range(100)]
h_n = [(1 / 2) ** i for i in range(100)]
n_0_4 = numpy.zeros(100)
n_0_4[0:5] = 1

y = signal.convolve(n_0_4, h_n)
y[5:] = 0

pyplot.stem(n, y[0:100])
pyplot.xlabel('n')
pyplot.ylabel('convolution')
pyplot.title('0 <= n <= 4')
pyplot.grid()
pyplot.show()
```

$n > 4$

```python
y = signal.convolve(n_0_4, h_n)
y[0:5] = 0

pyplot.stem(n, y[0:100])
pyplot.xlabel('n')
pyplot.ylabel('convolution')
pyplot.title('n > 4')
pyplot.grid()
pyplot.show()
```

So the final result would be :

```python
y = signal.convolve(n_0_4, h_n)

pyplot.stem(n, y[0:100])
pyplot.xlabel('n')
pyplot.ylabel('convolution')
pyplot.grid()
pyplot.show()
```
