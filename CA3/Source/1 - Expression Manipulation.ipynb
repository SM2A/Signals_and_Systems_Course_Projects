# Expression Manipulation

```python
import lcapy

x = lcapy.expr('A_1 * exp(-alpha * t) * u(t) + A_2 * exp(-beta * t) * sin(omega_0 * t + theta) * u(t)')
x.expr
```

```python
X = x(lcapy.s)
X.expr
```

```python
X.partfrac().simplify().expr
```

```python
X.partfrac().simplify_terms().expr
```

```python
X.ZPK().expr
```
