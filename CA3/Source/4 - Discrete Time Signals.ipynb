# Discrete Time Signals

## Discrete Analysis


```python
from lcapy import n, j, exp, u, pi

x_n = (5 * exp((-1 * j * n * pi) / 4) * u(n)) + (10 * exp((-1 * j * n * pi) / 3) * u(n))
x_n.plot((0,30))
```

```python
x_n.plot((0, 30), figsize=(8, 8), polar=True)
```

## Sequences

```python
x_1 = exp((-1 * j * n * pi) / 2) * u(n)
x_1_seq = x_1.seq((0,5))
x_1_seq
```

```python
x_2 = n * u(n)
x_2_seq = x_2.seq((0,5))
x_2_seq
```

```python
x_3 = x_1_seq + x_1_seq
x_3
```
