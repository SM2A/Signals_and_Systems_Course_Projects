# Discrete Time Signals

## Discrete Analysis


```python
from lcapy import n, j, exp, u, pi

x_n = (5 * exp((-1 * j * n * pi) / 4) * u(n)) + (10 * exp((-1 * j * n * pi) / 3) * u(n))
x_n.plot((0,30))
```

```python
x_n.plot((0, 30), figsize=(8, 8), polar=True)
```

## Sequences

```python
x_1 = exp((-1 * j * n * pi) / 2) * u(n)
x_1_seq = x_1.seq((0,5))
x_1_seq
```

```python
x_2 = n * u(n)
x_2_seq = x_2.seq((0,5))
x_2_seq
```

```python
x_3 = x_1_seq + x_1_seq
x_3
```

## Discrete Transform

```python
import lcapy

x_n = lcapy.expr('((0.5 - alpha) ** n) * u(n) + 10 * ((alpha + 1/3) ** n) * u(n)')
x_n.expr
```

```python
X = x_n.subs({'alpha':1})
Y = X(lcapy.z)
Y.plot()
```

```python
X = x_n.subs({'alpha':1.5})
Y = X(lcapy.z)
Y.plot()
```

```python
X = x_n.subs({'alpha':3})
Y = X(lcapy.z)
Y.plot()
```

```python
x_n = lcapy.expr('n * (alpha ** n) * u(n)')
x_n.expr
```

```python
X = x_n(lcapy.z)
X.expr
```

```python
x = X(n)
x.simplify_terms().expr
```

```python
x_dft = x_n.DFT()
x_dft.expr
```

```python
x_dtft = x_n.DTFT()
x_dtft.expr
```

```python
x_dftt_f = x_n.DTFT(lcapy.F)
x_dftt_f.expr
```

```python
x_dftt_o = x_n.DTFT(lcapy.Omega)
x_dftt_o.expr
```

```python
x_dftt_o = x_dftt_o.subs({'alpha':3})
x_dftt_o.plot()
```
